```python
data = pd.read_csv('data_rok.csv')

themes = [
    "Výzva k odstranění vozidla",
    "Změny rozpočtu",
    "nedoplatek",
    "Záměr pronájmu",
    'záměr prodeje',
    "Záměr pachtu částí pozemků"
]

result_data = pd.DataFrame()

for city in data['město'].unique():
    city_data = data[data['město'] == city].copy()
    
    for theme in themes:
        city_data[theme] = city_data['nazev'].str.contains(theme, case=False, na=False).astype(int)
    
    grouped_city_data = city_data.groupby(['město', 'rok'])[themes].sum().reset_index()
    
    result_data = pd.concat([result_data, grouped_city_data], ignore_index=True)
    
result_data
```

```python
data = pd.read_csv('all_data.csv',delimiter= ';')
data.head()
```

```python
data['datum'] = pd.to_datetime(data['datum'], errors='coerce')
data['rok'] = data['datum'].dt.year.astype('Int64')

data.head()
```

```python
from collections import Counter
import pandas as pd
import re

data['nazev'] = data['nazev'].str.lower()


def get_bigrams(text):
    
    words = re.findall(r'\b\w+\b', text)
    bigrams = zip(words, words[1:])
    return [' '.join(bigram) for bigram in bigrams]

all_text = ' '.join(data['nazev'].dropna())

all_bigrams = get_bigrams(all_text)

bigram_counts = Counter(all_bigrams)

bigram_counts_df = pd.DataFrame(bigram_counts.items(), columns=['Bigram', 'Count']).sort_values(by='Count', ascending=False)

print(bigram_counts_df.head(50))
```

```python
themes = [
    # Trendy v dotacích a financování:
    'poskytnutí dotace',
    'investiční dotace',
    # Legislativní a právní záležitosti:
    'veřejnoprávní smlouva',
    '106/1999',
    # Městské plánování a infrastruktura:
    "úpravy provozu",
    "přechodné úpravy",
    'místní úpravy',
    # Veřejné oznámení a vyhlášky:
    'veřejná vyhláška',
    'dražební vyhláška',
    'opatření obecné povahy',
    # Energetika a služby:
    "dodávky elektrické energie"
]

result_data = pd.DataFrame()

for city in data['město'].unique():
    city_data = data[data['město'] == city].copy()
    
    for theme in themes:
        city_data[theme] = city_data['nazev'].str.contains(theme, case=False, na=False).astype(int)
    
    grouped_city_data = city_data.groupby(['město', 'rok'])[themes].sum().reset_index()
    
    result_data = pd.concat([result_data, grouped_city_data], ignore_index=True)
    
result_data
```
